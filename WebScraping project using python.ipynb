```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://en.wikipedia.org/wiki/List_of_largest_technology_companies_by_revenue"

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"
}
```

```python
response = requests.get(url, headers=headers)
```

```python
soup = BeautifulSoup(response.text, "html.parser")
```

```python
tables = soup.find_all("table", class_="wikitable")
len(tables)
```

```python
main_table = tables[0]
```

```python
main_table = tables[1]
```

```python
columns = [th.text.strip() for th in main_table.find_all("th")]
columns
```

```python
df = pd.DataFrame(columns=columns)
```

```python
rows = main_table.find_all("tr")[1:]  # header skip

for row in rows:
    row_data = row.find_all("td")
    row_text = [data.text.strip() for data in row_data]
    df.loc[len(df)] = row_text
```
